# Model Comparison

```python
bestModels.sort_values(by=["meanAccuracy"], ascending=False, inplace=True)
bestModels.reset_index(drop=True, inplace=True)
bestModels
```

```python
plt.figure(figsize=(12, 11))

plt.scatter(bestModels["timeElapsed"], bestModels["meanAccuracy"])

plt.xlim(0, 10000)
plt.xticks(np.arange(0, 10000, 1000))
plt.ylim(0.9, 0.98)
plt.xlabel("Time Elapsed (seconds)")
plt.ylabel("Mean Accuracy")
plt.title("Model Accuracy vs. Time Elapsed")

for i, row in bestModels.iterrows():
    plt.annotate(row["modelName"],
                 (row["timeElapsed"], row["meanAccuracy"]),
                 textcoords="offset points",
                 xytext=(5,5))

plt.grid(True)
plt.show()
```
